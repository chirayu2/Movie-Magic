```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
#movies.head(1)
movies.shape
credits.shape
```

```python
credits.head(1)
```

```python
#we donot need two dataset sowe merge the dataset into one by using title
movies=movies.merge(credits,on='title')
```

```python
movies.shape # while merging always check if the shape isnt changed
```

```python
movies.info()
movies.isnull().sum() #check no of null values
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']] # selected required features
```

```python
import ast
movies.isnull().sum()
movies.dropna(inplace=True) #inplace true represents orignal data to be manipulated its important to drop all null coloumns or atleast manage it
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert) #conver to array
movies['genres']
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    y=0;
    for i in ast.literal_eval(obj):
        if(y!=3):
            l.append(i["name"])
            y=y+1
        else:
            break

    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)#selected top 3 cast
```

```python
movies['overview']=movies['overview'].apply(lambda p:p.split())
```

```python
movies
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i["name"])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')# remove english words and select 5000 features
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
 cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()# similar text removed
```

```python
def stem(text):
    #remove similar kind of text using a library called nltk like loves loved loving
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('loved loves loving')#example all are converted to love
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()#create new vectors after applying the stem fnction
```

```python
cv.get_feature_names()#you can see the changes in the features
```

```python
from sklearn.metrics.pairwise import cosine_similarity # bog similarity bw movies
```

```python
cosine_similarity(vectors).shape

'''similarity matrix is of 4806x4806 shape because each
   movies is compared with another movie
'''
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```
